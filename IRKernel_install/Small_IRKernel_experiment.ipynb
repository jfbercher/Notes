## A small experiment with linear regression
- lm works,
- plots work
- tab completion work (but not shift-TAB unfortunately)
- help, ? work
- this is quite nice

```r
x=rnorm(n = 100)
t=0:99
y=0.05*t+x
plot(y, pch = 19, col = "blue", xlab = "Time", sub = "(seconds)", main = "Linear regression" )
#plot(y, type = 'l', col = "red", xlab = "Time", sub = "(seconds)" )
z=lm(y~t)
print("Linear regression on the little thing")
print(summary(z))
abline(z)
```
